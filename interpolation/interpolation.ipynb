```python
from scipy import linalg
import numpy as np
import math
import matplotlib.pyplot as plt

def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)
a = 1
b = 15
```

```python
def compute_polynomial(f, x):
    n = len(x)
    b = np.vectorize(f)(x)
    A = np.ndarray((n, n))
    for i in range(n):
        A[i] = np.power(x, i)
    A = A.T
    w = linalg.solve(A, b)
    return np.poly1d(w[::-1])

x = np.arange(a, b, 0.1)
y = np.vectorize(f)(x)
plt.plot(x, y)

Xs = [
    [1, 15],
    [1, 8, 15],
    [1, 4, 10, 15]
]
    
for X in Xs:
    poly = compute_polynomial(f, X)
    plt.plot(x, poly(x))
```
